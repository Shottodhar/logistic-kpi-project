```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
drop_cols = [
    'Customer Email',
    'Customer Fname',
    'Customer Lname',
    'Customer Password',
    'Customer Street',
    'Product Image'
]

df = df.drop(columns=drop_cols)
```

```python
df
```

```python
df['order date (DateOrders)'] = pd.to_datetime(
    df['order date (DateOrders)']
)

df['shipping date (DateOrders)'] = pd.to_datetime(
    df['shipping date (DateOrders)']
)
```

```python
df.info()
```

```python
df['Delivery Performance'] = np.where(
    df['Late_delivery_risk'] == 1,
    'Late',
    'On Time'
)
```

```python
df
```

```python
df['Profit Margin %'] = (
    df['Order Profit Per Order']
    / df['Sales']
) * 100
```

```python
df['Order Year'] = df['order date (DateOrders)'].dt.year
```

```python
df['Order Month'] = df['order date (DateOrders)'].dt.month_name()
```

```python
df['Order Quarter'] = df['order date (DateOrders)'].dt.quarter
```

```python
df
```

```python
df[['Delivery Performance',
    'Profit Margin %',
    'Order Year',
    'Order Month',
    'Order Quarter']].head()
```

```python
df.to_csv(
    "../data/processed/logistics_cleaned.csv",
    index=False
)
```

```python
#Revenue Analysis
#1. Which region Generates the Highest Revenue?
region_revenue = (
    df.groupby('Order Region')['Sales']
    .sum()
    .sort_values(ascending=False)
)

region_revenue.head(10)
```

```python
plt.figure(figsize=(12,5))

region_revenue.head(10).plot(kind='bar')

plt.title('Top 10 Revenue Regions')

plt.show()
```

```python
#2. Which markets Generate the Highest Revenue?
market_revenue = (
    df.groupby('Market')['Sales']
    .sum()
    .sort_values(ascending=False)
)

market_revenue
```

```python
plt.figure(figsize=(10,5))

market_revenue.plot(kind='bar')

plt.title('Revenue by Market')

plt.show()
```

```python
#Profit Analysis
#which region generate profit
region_profit = (
    df.groupby('Order Region')
    ['Order Profit Per Order']
    .sum()
    .sort_values(ascending=False)
)

region_profit.head(10)
```

```python
plt.figure(figsize=(12,5))

region_profit.head(10).plot(kind='bar')

plt.title('Top 10 Profitable Regions')

plt.show()
```

```python
#which category generate profit?
category_profit = (
    df.groupby('Category Name')
    ['Order Profit Per Order']
    .sum()
    .sort_values(ascending=False)
)

category_profit
```

```python
plt.figure(figsize=(10,5))

category_profit.plot(kind='bar')

plt.title('Profit by Category')

plt.show()
```

```python
#Delivery Analysis
#late delivery rate:
df['Delivery Performance'].value_counts(normalize=True)*100
```

```python
#6. Which Shipping Mode Has the Highest Late Delivery Rate?
shipping_delay = pd.crosstab(
    df['Shipping Mode'],
    df['Delivery Performance']
)

shipping_delay
```

```python
#percentage:
shipping_delay_pct = pd.crosstab(
    df['Shipping Mode'],
    df['Delivery Performance'],
    normalize='index'
)*100

shipping_delay_pct
```

```python
#Customer Segment Analysis
#7. Which Customer Segment Generates the Highest Revenue?
segment_sales = (
    df.groupby('Customer Segment')
    ['Sales']
    .sum()
    .sort_values(ascending=False)
)

segment_sales
```

```python
8. Which Customer Segment Generates the Highest Profit?
segment_profit = (
    df.groupby('Customer Segment')
    ['Order Profit Per Order']
    .sum()
    .sort_values(ascending=False)
)

segment_profit
```

```python
#Top 10 Product by Revenue
top_products = (
    df.groupby('Product Name')
    ['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```

```python
#Top 10 Product by Profit
top_profit_products = (
    df.groupby('Product Name')
    ['Order Profit Per Order']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_profit_products
```
